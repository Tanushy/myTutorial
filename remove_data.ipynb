```python

import pandas as pd
import numpy as np
df = pd.read_csv("./CSV-Files/Portfolio11.11.2022.csv")
df
```

```python

ind = 0
for item in df['Produkt']:
    tmp = item.split(' ')
    if tmp[0] == 'ISHARES':
        df.drop(index=ind, inplace=True)
    if tmp[0] == 'ODAX':
        df.drop(index=ind, inplace=True)
    if tmp[0] == 'ISHR':
        df.drop(index=ind, inplace=True)
    if tmp[0] == 'ANN':
        df.drop(index=ind, inplace=True)
    if tmp[0] == 'AS1':
        df.drop(index=ind, inplace=True)
    if tmp[0] == 'GENERAL':
        df.drop(index=ind, inplace=True)
    if tmp[0] == 'CASH':
        df.drop(index=ind, inplace=True)
    if tmp[0] == 'VANGUARD':
        df.drop(index=ind, inplace=True)
    ind += 1
df
```

```python
# Berechnung der Summe in EUR
summe = 0
for item in df['Wert in EUR']:
    summe += float(item.split(',')[0])
summe
```

```python
def remove_row(dataset, name):
    ind = 0
    for item in dataset['Produkt']:
        if item == name:
            dataset.drop(index=ind, inplace=True)
        ind += 1
```

```python
remove_row(df, 'ADESSO AG')
```

```python
remove_row(df, '2G ENERGY AG')
```
